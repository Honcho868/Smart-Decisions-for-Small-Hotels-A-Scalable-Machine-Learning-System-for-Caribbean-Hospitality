The customer_type column in the dataset categorizes guests based on the nature of their booking. It helps in understanding booking patterns and customer behavior. 

-Transient: Guests who book individual stays and do not belong to any group or contract. These bookings are usually short-term and can come from various sources like direct bookings, online travel agents, or walk-ins.

-Transient-party: Similar to transient customers, but they are traveling as part of a small group. They may have separate bookings but are associated with the same trip or purpose.

-Contract: Guests who book under a corporate or business contract. These bookings are usually made through a company or travel agency and often involve fixed rates for a specific period.

-Group: Bookings that are part of a large group, such as for conferences, weddings, or tour groups. These are typically arranged through travel agencies or event planners.

```python
# Identify and print columns with missing values
null_columns = df.isnull().sum()
null_columns = null_columns[null_columns > 0]
print(null_columns)
```

The dataset has 119390 rows and the column 'company' has 112593 missing values therefore it is being removed from the dataset due to a large quantity of missing data.

```python
# Drop 'company' column from the dataframe
df.drop(['company'], axis=1, inplace=True)
df.isnull().sum()
```

The agent column is being converted from numerical (agent ID) to categorical (agent or no agent) to simplify analysis by focusing on whether bookings involved an agent, rather than which specific agent was involved.

```python
# Convert the 'agent' column to indicate whether an agent was involved in the booking
df['agent'] = df['agent'].apply(lambda x: 'has agent' if not pd.isna(x) else 'no agent')
# Display the count of missing values across all columns
df.isnull().sum()
```

```python
df.dropna(subset=['children'], inplace=True)
df.isnull().sum()
```

```python
# Fill missing values in the 'country' column with the mode (most frequent value)
most_frequent_country = df['country'].mode()[0]
df['country'].fillna(most_frequent_country, inplace=True)
df.isnull().sum()
```

There were rows where the Adults, Children and Babies columns were all 0. This is not a logical situation since bookings cannot be made for zero guests and thus these rows were removed

```python
no_guests = df[(df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0)]
print("Rows where adults, children, and babies are all 0:", len(no_guests))
```

```python
df = df[~((df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0))] #Remove rows where Adults, children and babies are all zero

no_guests = df[(df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0)]
print("Rows where adults, children, and babies are all 0:", len(no_guests))
```

```python
# Remove rows with 'Undefined' in distribution_channel and meal columns
df = df[(df['distribution_channel'] != 'Undefined') & (df['meal'] != 'Undefined')]
```

```python
df
```

```python
df = df.drop(['name','email','phone-number','credit_card'], axis=1)
```

```python
df.info()
```

```python
#Change datatype for number of children from float to integer
df['children'] = df['children'].astype(int) 
```

An arrival date column will be created by merging the arrival year, month and day of month columns and converting the column tyoe to datetime. This is done to calculate the time difference (in days) between cancellation date (reservation status date) and the proposed arrival day

```python
# Convert month names to numbers
month_map = {
    'January': '01', 'February': '02', 'March': '03', 'April': '04',
    'May': '05', 'June': '06', 'July': '07', 'August': '08',
    'September': '09', 'October': '10', 'November': '11', 'December': '12'
}
df['arrival_date_month'] = df['arrival_date_month'].map(month_map)

# Create a new column by combining the date components
df['arrival_date'] = pd.to_datetime(df['arrival_date_year'].astype(str) + '-' + 
                                  df['arrival_date_month'].astype(str) + '-' + 
                                  df['arrival_date_day_of_month'].astype(str))
```

```python
df.info()
```

```python
# Convert reservation_status_date to datetime
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])

# Calculate the difference in days
df['days_before_arrival'] = (df['arrival_date'] - df['reservation_status_date']).dt.days

df
```

##### Dataset after preprocessing

```python
column_info = pd.DataFrame(columns=['Column', 'Type', 'Details'])

# Populate the DataFrame with column information
for col in df.columns:
    if col in ['hotel', 'arrival_date_month', 'meal', 'country', 'market_segment', 
               'distribution_channel', 'reserved_room_type', 'assigned_room_type', 
               'deposit_type', 'customer_type', 'reservation_status']:
        details = f"Categorical feature: {df[col].unique()}"
    elif col in ['lead_time', 'arrival_date_year', 'arrival_date_week_number', 
                'arrival_date_day_of_month', 'stays_in_weekend_nights', 
                'stays_in_week_nights', 'adults', 'children', 'babies', 
                'is_repeated_guest', 'previous_cancellations', 
                'previous_bookings_not_canceled', 'booking_changes', 
                'days_in_waiting_list', 'adr', 'required_car_parking_spaces', 
                'total_of_special_requests', 'days_before_arrival']:
        details = f"Numerical feature: Mean = {df[col].mean():.2f}, Std = {df[col].std():.2f}"
    elif col == 'is_canceled':
        details = f"Target feature: {df[col].unique()}"
    else:
        details = ""
    
    column_info = pd.concat([column_info, pd.DataFrame({
        'Column': [col],
        'Type': [df[col].dtype],
        'Details': [details]
    })], ignore_index=True)

# Display the DataFrame
print("Columns for df:")
column_info
```
